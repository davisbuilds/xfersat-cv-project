```python
# EuroSAT Dataset Exploration
# This notebook provides an overview of the EuroSAT dataset and visualizes examples from each class.

import os
import random
import matplotlib.pyplot as plt
from pathlib import Path
from PIL import Image
import numpy as np
```

```python
# Dataset parameters
DATASET_DIR = "../data/raw/EuroSAT_RGB"
```

```python
# Check if dataset exists
if not os.path.exists(DATASET_DIR):
    print("Dataset not found. Please check the path set for DATASET_DIR")
else:
    # Get all class directories
    class_dirs = [d for d in os.listdir(DATASET_DIR) if os.path.isdir(os.path.join(DATASET_DIR, d))]
    print(f"Found {len(class_dirs)} classes: {', '.join(class_dirs)}")
    
    # Plot examples from each class
    plt.figure(figsize=(15, 10))
    for i, class_name in enumerate(class_dirs):
        class_path = os.path.join(DATASET_DIR, class_name)
        image_files = os.listdir(class_path)
        random_image = random.choice(image_files)
        image = Image.open(os.path.join(class_path, random_image))
        
        plt.subplot(3, 4, i+1)
        plt.imshow(image)
        plt.title(class_name)
        plt.axis('off')
    
    plt.tight_layout()
    plt.show()
    
```

```python
# Calculate dataset statistics
class_counts = {}
for class_name in class_dirs:
    class_path = os.path.join(DATASET_DIR, class_name)
    class_counts[class_name] = len(os.listdir(class_path))

print("Images per class:\n")
for class_name, count in class_counts.items():
    print(f"{class_name}: {count} images")

print(f"\nTotal images: {sum(class_counts.values())}")
```
